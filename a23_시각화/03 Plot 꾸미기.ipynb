# Plot 꾸미기

```python
plt.plot(np.random.randn(50), color='g', marker='o', linestyle='--')
```

```python
# plot 꾸미기 옵션

# color
# 값 색상
# "b" blue
# "g" green
# "r" red
# "c" cyan
# "m" magenta
# "y" yellow
# "k" black
# "w" white

# marker
# 값 마킹
# "." point
# "," pixel
# "o" circle
# "v" triangle_down
# "^" triangle_up
# "<" triangle_left
# ">" triangle_right
# "8" octagon
# "s" square
# "p" pentagon
# "*" star
# "h" hexagon
# "+" plus
# "x" x
# "D" diamond

# line style
# 값 라인 스타일
# "-" solid line
# "--" dashed line
# "-." dash-dotted line
# ":" dotted line
# "None" draw nothing
```

```python
# 굳이 스타일에 대해 일일히 입력하지 않고, 아래와 같이 연속적으로 넘겨주어도 된다.
plt.plot(np.random.randn(30), 'k.-')
```

## plot 의 눈금, 레이블, 범위 지정

```python
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum())
```

```python
fig
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
ax.set_xticks([0, 250, 500, 750, 1000])
```

```python
fig = plt.figure(figsize=(8, 8))  # figure size (맨 처음에 주어야 한다)
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())

# x축 눈금 설정
ax.set_xticks([0, 250, 500, 750, 1000])

# 눈금을 문자로 바꾸기
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')

# 제목입력
ax.set_title('random walk plot')

# 레이블 입력
ax.set_xlabel('Stages')
ax.set_ylabel('Values')

# 눈금
ax.grid(True)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1, 1)
ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'b--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'r.', label='three')

# 범례 (legend)
ax.legend(loc="best")

# x축, y축 범위 수정
ax.set_xlim([100, 900])
ax.set_ylim([-100, 100])
```
