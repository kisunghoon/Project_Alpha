wordcloud 라이브러리는 matplotlib 과 함께 사용됨

```python
import wordcloud
import matplotlib.pyplot as plt
```

```python
wordcloud.__version__
```

```python
words = {
    'Python' : 10,
    'Java' : 5,
    'C' : 7,
    'C++': 9,
    'JavaScript': 8
}
```

```python
wc = wordcloud.WordCloud()
wc
```

```python
cloud = wc.generate_from_frequencies(words)
cloud
```

```python
plt.imshow(cloud)
```
