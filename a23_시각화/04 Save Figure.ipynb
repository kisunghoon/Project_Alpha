# 그래프 <--> 파일
파일로 저장 : savefig()

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import font_manager, rc
import platform

try : 
    if platform.system() == 'Windows':
    # 윈도우인 경우
        font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
        rc('font', family=font_name)
    else:    
    # Mac 인 경우
        rc('font', family='AppleGothic')
except : 
    pass
matplotlib.rcParams['axes.unicode_minus'] = False  

```

```python
import os

if not os.path.exists('out'):
    os.mkdir('out')
```

```python
x = [1, 2, 3]
y = [10, 20, 30]
plt.plot(x, y, color="Green", linestyle="-", marker="v")
plt.title("3rd Green Graph")

plt.savefig(r'.\out\figimage1.png')

 # 지원포맷 eps, pdf, pgf, png, ps, raw, rgba, svg, svgz
```

### 이미지 저장시 해상도 (크기) 조정
savefig(dpi = )

```python
x = [1, 2, 3]
y = [10, 20, 30]
plt.plot(x, y, color="Green", linestyle="-", marker="v")
plt.title("3rd Green Graph")

plt.savefig(r'.\out\figimage1.png', dpi=100)  # dpi= 이미지 해상도 
plt.imread(r'.\out\figimage1.png').shape  # 이미지 읽어들어와서 해상도 확인해보기
```

## 벡터 포맷 이미지 저장

```python
x = [1, 2, 3]
y = [10, 20, 30]
plt.plot(x, y, color="Green", linestyle="-", marker="v")
plt.title("3번째 그래프")

plt.savefig('./out/figimage1.svg')
```
