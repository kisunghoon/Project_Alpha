# urllib
- 파이썬 표준모듈로써 URL을 다루이 위한 모듈 패키지
- 설치가 필요하지 않음 
- requests 모듈과 마찬가지로 URL과 관련된 다양한 기능 제공 

```python
import urllib

URL = 'https://www.naver.com'
response = urllib.request.urlopen(URL)
data = response.read()
print(data.decode('utf-8'))
```

```python
import requests

URL = 'https://www.naver.com'
response = requests.get(URL)
data = response.text
print(data)
```

## urllib 의 중요 모듈
1. request
    - URL을 열고 읽는 모듈 (HTTP request)
1. error
    - request 모듈에서 발생하는 에러들을 포함하는 모듈
1. parse
    - URL을 파싱하는 모듈 (URL 해석및 조작)
1. robotparser
    - robots.txt 파일을 파싱하는 모듈

### urllib.request.Request()
- 해당 URL 요청 객체 생성, HTTPRequest 객체 리턴
- 여러가지 함수들 제공: 전체 URL, 프로토콜 타입, 호스트 ...

```python
request = urllib.request.Request(URL)

print(request.full_url)
print(request.type)
print(request.host)
```

### urllib.request.urlopen()
- 해당 URL 열기, 응답데이터는 바이트형식(바이너리) 의 HTTPResponse 객체
- request 객체 또는 URL 을 직접 넣어도 가능

```python
urllib.request.urlopen(URL)
```

```python
urllib.request.urlopen(request)
```

### urllib.requet.read()
- urlopen() 으로 연 객체를 읽고, 인자로 전달되는 숫자만큼 데이터를 읽음 -> 바이트형식
- readlines() : 페이지 데이터(바이트 형식) 를 줄 단위로 읽어 리스트로 리턴

```python
response = urllib.request.urlopen(URL)

# response.read()  # 매개변수 없으면 response 전체 바이트 데이터 리턴

response.read(300)
```

```python
response = urllib.request.urlopen(URL)

response.readlines()  # 바이트 데이터의 리스트 리턴
```

### decode()

- 바이트형식의 데이터를 원하는 형식으로 변환
- 기본값 utf-8

```python
response = urllib.request.urlopen(URL)
data = response.read(500)
data.decode()
```
