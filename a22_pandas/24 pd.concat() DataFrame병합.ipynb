#  pd.concat()
DataFrame 병합시키기

```python
import pandas as pd
import numpy as np
```

## pd.concat() 함수 사용하여 dataframe 병합하기
 - pandas.concat 함수
 - 축(axis)을 따라 dataframe을 병합 가능
   - axis = 0 → 행단위 병합 (디폴트)
   - axis = 1 → 열단위 병합

## column 명이 같은 경우

```python
df1 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
df2 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
```

```python
df1
```

```python
df2
```

```python
# concat( DataFrame 리스트 ) <-- 행(row) 단위로 병합 (axis=0)
pd.concat([df1, df2])

# index 가 반복되는것도 확인 가능
```

```python
# 만약 인덱스를 무시하고 싶다면!
pd.concat([df1, df2], ignore_index=True)

# 새로운 0-base 인덱스가 생성된 것을 확인할수 있다.
```

```python
# 열 level 로 concat 하기 (axis=1)

pd.concat([df1, df2], axis=1)
```

## column 명이 다른 경우

```python
df3 = pd.DataFrame({'key2' : np.arange(10), 'value2' : np.random.randn(10)})
df3  # 컬럼명이 df1, df2 와 다르다!
```

```python
pd.concat([df1, df3])

# 서로 없는 컬럼값은 NaN 으로 처리됨.
```

```python
pd.concat([df1, df3], axis=1)
```

### 위에서 서로 다른 index 끼리 열방향 (axis=1) concat 하면?

```python
df2.index += 10
df2
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2])
```
