## NaN 처리 방법
 - 데이터에서 삭제

 - 다른 값으로 치환

#### NaN 데이터 삭제하기

```python
df.dropna()

# 183개의 row 가 나온다 (확 줄었다)
# PassengerId 를 보면 중간중간에 빠진거 확인


# 기본적으로 row 기준(axis=0)으로 동작한다
# 각 row 에서, 하나라도 NaN 값이 있으면 해당 row 를 지워버린다
```

```python
# 특정 칼럼에 대해서 적용할 경우
# Age 칼럼이 NaN 인 경우만 drop 할 경우

df.dropna(subset = ['Age'])
```

```python
df.dropna(subset = ['Age', 'Cabin'])
```

```python
# '컬럼' 에 NaN 이 있으면 '컬럼' 을 삭제하기
df.dropna(axis = 1)

# Age, Cabin, Embardked 컬럼이 삭제됨!
```

### NaN 값 대체하기
 - 평균으로 대체하기
 - 생존자/사망자 별 평균으로 대체하기

```python
# 평균값
df['Age'].mean()  # NaN 은 배제된 연산 결과
```

```python
df['Age'].sum() / df['Age'].count()
```

```python
# NaN 을 평균값으로 대체하기 fillna()
df['Age'].fillna(df['Age'].mean())
```

### 도전] Age NaN 값을 생존자/ 사망자 별 평균으로 대체 하기

```python
# 생존자 나이 평균 구하기
mean1 = df[df['Survived'] == 1]['Age'].mean()
mean1
```

```python
# 사망자 나이 평균 구하기
mean0 = df[df['Survived'] == 0]['Age'].mean()
mean0
```

```python
# 생존자 중 Age 값이 NaN 인 사람을 fillna(생존자 나이평균)  하기
df[df['Survived'] == 1]['Age'].fillna(mean1)  
```

```python
# 사망자 중 Age 값이 NaN 인 사람을 fillna(사망자 나이평균)  하기
df[df['Survived'] == 0]['Age'].fillna(mean0)
```

```python
# 실제 데이터에 넣을때 inplcae=True 대신 이렇게 해도 된다.
```

```python
df.loc[df['Survived'] == 1, 'Age'] = df[df['Survived'] == 1]['Age'].fillna(mean1)
```

```python
df.loc[df['Survived'] == 0, 'Age'] = df[df['Survived'] == 0]['Age'].fillna(mean0)
```
