### world DB 의 내용을 csv 로 저장
저장경로: mysql/world 

```python
# world DB 의 내용을 -> csv 로 저장
dbname = "world"

db = MySQLdb.connect(
    "localhost",  # DATABASE_HOST
    "root",  # DATABASE_USERNAME
    "1234",  # DATABASE_PASSWORD
    dbname,  # DATABASE_NAME
    charset='utf8',
)

path = os.path.join(dataset_path, "mysql", dbname)
if not os.path.exists(path):
    print(path, '를 생성합니다')
    os.makedirs(path)

table_names = pd.read_sql("SHOW TABLES;", db)[f"Tables_in_{dbname}"]

for table_name in table_names:
    # 1. SQL Query 를 동적으로 만들어주고,
    SQL_QUERY = "SELECT * FROM {};".format(table_name)
    # 2. 만들어진 SQL Query 로 DataFrame 을 불러오고,
    table_df = pd.read_sql(SQL_QUERY, db)
    
    # 3. 저장하기 
    file_path = os.path.join(dataset_path, "mysql", dbname, f"{table_name}.csv")
#     print(file_path) # 확인
    
    table_df.to_csv(file_path, index=False, encoding='utf8')
    print(f"{file_path} 생성 ")

db.close()
```
