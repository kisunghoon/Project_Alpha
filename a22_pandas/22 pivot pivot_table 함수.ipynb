 - dataframe의 형태를 변경
 - 인덱스, 컬럼, 데이터로 사용할 컬럼을 명시

```python
#  pivot(index=None, columns=None, values=None)

# index 로 지정할 컬럼
# column 으로 지정할 컬럼
# values 나머지 채울 값들.
```

```python
# 이제 지역별로(index)  요일별 강수량/강수확률을 보고싶다
```

```python
df.pivot('지역', '요일')  # index="지역", columns ='요일'
```

```python
# ※ 만약 
# df.pivot('지역', '요일')  # <-- '지역', '요일' 이 같은 데이터가 있으면 에러 난다
```

# pd.pivot_table()
 - 기능적으로 pivot과 동일
 - pivot과의 차이점
   - 중복되는 모호한 값이 있을 경우, aggregation 함수 사용하여 값을 채움

```python
pd.pivot_table(df, index='지역', columns='요일')   # 기존의 pivot() 과 거의 유사
```

```python
# pivot() 과의 차이점은 중복데이터의 처리수단 제공여부다.
```

```python
# 중복데이터를 만들어 본다.  서울-월요일
df2 = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '월요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
                  })

df2  # 서울 - 월요일 <-- 2개 있다.
```

```python
# df2.pivot('지역', '요일')  # 에러
```

```python
pd.pivot_table(df2, index='지역', columns='요일', aggfunc=np.mean)  # 서울-월요일 은 '평균값' 으로 채운다
```

```python
# aggfunc=np.mean  <-- default 값이다
df3 = pd.pivot_table(df2, index='지역', columns='요일')
df3
```
