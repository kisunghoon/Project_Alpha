## index를 이용한 group by
 - index가 있는 경우, groupby 함수에 level 사용 가능
   - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가
   
* **set_index** 함수
 - column 데이터를 index 레벨로 변경
* **reset_index** 함수
 - 인덱스 초기화

```python
df.head()  # 현재는 0-base 인덱스로 붙어 있슴
```

```python
# 인덱스를 바꾸려면 set_index()
```

```python
df.set_index('Pclass')  # 원본 변화 없슴
```

```python
# multi index
df.set_index(['Pclass', 'Sex'])
```

```python
df.set_index(['Pclass', 'Sex']).reset_index()

# 기존의 index 로 내려간 것들이 다시 컬럼으로 올라오고
# 0-base 인덱스가 새로이 붙음.
```

```python
# index 가 있는 DataFrame 을 groupby() 할때
# level 이나 함수를 이용할수 있다.
```

```python
df.set_index('Age').groupby(level=0).mean()

# level=0 : 0번째 level 의 인덱스 (이 경우 Age) 기준으로 groupby 발생
```

```python
# 함수를 사용하여 그룹핑
# 나이대별 생존률 구하기
```

```python
df.set_index('Age')
```

```python
# 함수 만들기
import math
def age_cagegorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10   # <-- 이 리턴하는 값이 곧 gouping 기준이 된다.
```

```python
# 위 함수를 groupby(함수) 에 적용 : 인덱스로 그룹핑
df.set_index('Age').groupby(age_cagegorize).mean()
```

```python
df.set_index('Age').groupby(age_cagegorize).mean()['Survived']
```

## aggregate(집계) 함수 사용하기
 - groupby 결과에 '집계함수'를 적용하여 그룹별 데이터 확인 가능

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).aggregate(np.sum)
```

```python
# 여러개의 집계함수를 list 형태로 건네줄수 있다.

df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).aggregate([np.sum, np.mean, np.max])

# multi-level columns !
```

```python
df2 = df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).aggregate([np.sum, np.mean, np.max])
df2
```

#### groupby 객체는 iterable 하다!

```python
for groupid, group_df in df.groupby('Pclass'):
    print(groupid, len(group_df))
```
