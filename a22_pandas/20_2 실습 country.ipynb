## 하나의 파이썬 코드로 만들어 보세요

```python
import numpy as np
import pandas as pd
import os

filepath = r'D:\DevRoot\DataSet\mysql\world\country.csv'
df_country = pd.read_csv(filepath)
df_asia = df_country[df_country.Continent == "Asia"]  # 아시아 국가들
kor_population = df_country[df_country.Name == "South Korea"].Population.iloc[0]  # 한국 인구
df_result = df_asia[df_asia.Population > kor_population][["Name", "Population"]] 
df_result = df_result.sort_values(["Population"], ascending=[False])
df_result = df_result.reset_index(drop=True)
df_result.index = df_result.index + 1
if not os.path.exists('out'):
    os.makedirs('out')
    print('out', '디렉토리 생성')
df_result.to_csv("./out/result_2.csv")
print('완료')
```

#### inplace = True
원본 변화 발생

```python
import numpy as np
import pandas as pd
import os

filepath = r'D:\DevRoot\DataSet\mysql\world\country.csv'
df_country = pd.read_csv(filepath)
df_asia = df_country[df_country.Continent == "Asia"]  # 아시아 국가들
kor_population = df_country[df_country.Name == "South Korea"].Population.iloc[0]  # 한국 인구
df_result = df_asia[df_asia.Population > kor_population][["Name", "Population"]] 
df_result.sort_values(["Population"], ascending=[False], inplace=True) # ★원본변화
df_result.reset_index(drop=True, inplace=True)  # ★원본변화
df_result.index += 1
if not os.path.exists('out'):
    os.makedirs('out')
    print('out', '디렉토리 생성')
df_result.to_csv("./out/result_2.csv")
print('완료')
```

# 남미대륙 (South America) 소속국가들중 평균면적보다 작은 나라들은?
- '이름' 과 '면적' 출력
- 면적이 작은것부터 오름차순으로 정렬
- index 는 1부터 시작
- ./out/country_3.csv 파일로 저장

```python
import numpy as np
import pandas as pd

filepath = r'D:\DevRoot\DataSet\mysql\world\country.csv'
df_country = pd.read_csv(filepath)
df_south = df_country[df_country.Continent == "South America"] 
df_result = df_south[df_south.SurfaceArea < df_south.SurfaceArea.mean()][["Name", "SurfaceArea"]]
df_result.sort_values(["SurfaceArea"], inplace=True)
df_result.reset_index(drop=True, inplace=True)
df_result.index += 1
df_result.to_csv("./out/country_3.csv")
```
