## row 선택하기
  - Series의 경우 []로 row 선택이 가능하나, **DataFrame의 경우는 기본적으로 column을 선택하도록 설계**
  - **.loc, .iloc함수**로 row 선택 가능
    - loc - 인덱스 자체를 사용
    - iloc - 0 based index로 사용
    - 이 두 함수는 ,를 사용하여 column 선택도 가능

```python
train_data.head()  # csv 읽어올때, 특별히 지정 안하면 0 base index 가 붙여짐
```

```python
# .loc[] 을 사용하여  row 하나 선택
train_data.loc[986]  # 결과 Series
```

```python
# .loc[] 을 사용하여  row 여러개 선택
train_data.loc[[986, 100, 100, 990]]   # DataFrame 결과
```

```python
train_data.iloc[[0, 100, 200, 2]]  # 여러 row 가능
```

```python
train_data.loc[[0, 100, 200, 2]] 
```

```python
train_data.iloc[:5]
```

## row, column 동시에 선택하기
loc, iloc 속성을 이용할때, 콤마를 이용하여 둘다 명시 가능

```python
# loc[row index, column]
train_data.loc[[986, 110, 110, 990], ['Survived', 'Name', 'Sex', 'Age']]
```

```python
# 동일한 결과!
train_data[['Survived', 'Name', 'Sex', 'Age']].loc[[986, 110, 110, 990]]
```

```python
# 동일한 결과!
train_data.loc[[986, 110, 110, 990]][['Survived', 'Name', 'Sex', 'Age']]
```

```python
# 데이터 프로그래밍 하다보면, row -> column 순으로 접근하다가도
#  column -> row 순으로 접근하기도 한다.
#  pandas 를 사용하면 어떠한 순서로도 접근이 용이하다는 뜻이다
# 익숙해지면, 걍 의식의 흐름(?) 대로 데이터 프로그래밍이 가능해진다!! 옷홍!
```
