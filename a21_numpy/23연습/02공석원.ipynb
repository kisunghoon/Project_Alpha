```python
# 1번 유저의 '평점 개수' 와 '평점 평균' 을 구하세요
```

```python
user_rating_counting = df.groupby(by=['userid']).count()
user_rating_counting.drop(columns=["movieid"])
user_rating_counting
```

```python
total_rating_table = df.groupby(by=['userid']).sum().groupby(level=[0]).cumsum()
total_rating_table.drop(['movieid'],axis=1)
total_rating_table
```

```python
user_rating_avg = total_rating_table/user_rating_counting
user_rating_avg
```

```python
result = pd.concat([user_rating_counting,user_rating_avg],axis=1)
result
```
