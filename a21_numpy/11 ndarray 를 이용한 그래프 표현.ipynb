# numpy 시각화 (visualization)
- matploblib 사용

 1. matplotlib 모듈 이용하여 그래프 표현하기
 
 > `맷플롯립, 맷플롯라이브러리` 라고들 발음합니다

- 데이터 시각화는 데이터 분석, 머신러닝에서 매주 중요한 프로세스입니다.
- 시각화를 통해 데이터 추세나 분포 등에 대한 인사이트와 전략을 세울수 있도록 해야 합니다

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline   
# jupyter notebook 에 바로 그려질수 있는 notebook 기능
```

```python
# 데이터 준비
```

```python
x = np.linspace(0, 10, 11)
x
```

```python
y = x ** 2 + x + 2
y
```

### 그래프 출력하기
 - plot함수 (선 그래프), scatter(점 그래프), hist(히스토그램) 등 사용
   - 함수의 parameter 혹은 plt의 다른 함수로 그래프 형태 및 설정을 변경 가능
   - 기본적으로, x, y에 해당하는 값이 필요

```python
plt.plot(x, y)  # 선 그래프
```

```python
plt.scatter(x, y)
```

```python
y = x ** 2 + x + 2 + np.random.randn(11)
y
```

```python
plt.plot(x, y)
```

## 그래프에 레이블, 타이틀 추가 
- xlabel(), ylabel(), title()

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.plot(x, y)
```

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)  # 눈금 추가
plt.plot(x, y)
```

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)
plt.xlim(0, 20)  # x, y축 표시범위 지정
plt.ylim(0, 200)
plt.plot(x, y)
```
