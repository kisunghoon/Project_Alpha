# Cron
- 유닉스 계열 (Linux, Mac OS 등) OS 에서 시간을 기반으로 한 **작업 예약 스케쥴러**
    - 별도의 설치 없이 기본으로 사용 가능
- 주기적으로 거정도니 시간, 날짜, 간격에 원하는 프로그램을 실행할수 있게 함
- 서버는 꺼지지 않고 항상 작동 중인 것을 활용하기에, 주로 서버에 적용
    - 평소에는 실행되지 않고 대기하다가 특정 시간이 되면 해당 프로그램을 실행
   
- 명령어 중심

### Cron 방식 스케쥴 기술
- 프로그램 실행을 시간 형식에 맞게 예약

- 시간형식
    - Minute(0 ~ 59) : 분
    - Hour (0 ~ 23) : 시간
    - Day (1 ~ 31) : 일
    - Month (1 ~ 12) : 월
    - Week (0 ~ 6) : 요일, 0 을 일요일
    
- 예시
    > **0 12 1 1 3**<br>    
    >> ↑ 1월 1일 수요일 12시 정각에 실행
    
    
- 일정 간격의 모든 시간을 지정할 경우 * 사용
    > 30 * * * *<br>
    >> ↑ 매시간 30분 마다 실행

- 일정 시간 간격을 반복하도록 지정할 경우 , 사용
    > 10,20,30 * * * *<br>
    >> ↑ 매시간 10분, 20분, 30분에 실행

- 시간 범위를 지정할 경우 - 사용
    > 30 13-15 * * * <br>
    >> ↑ 13시, 14시, 15시 30분에 실행

- 시간 간격을 지정할 경우 / 사용
    > 0-30/5 * * * * <br>
    >> ↑ 매시간 0분부터 30분까지 5분 간격으로 실행

    
    

- 터미널창(콘솔)에서 **crontab -e** 명령어로 cron 설정 파일 편집
    - 현재 로그인 한 사용자에서 실행



- 작업스케줄 설정 + 실행할 명령어

- 예시

    > 10 * * * * python /Users/zoostar/Desktop/test.py
    
    >> ↑ 매 10분마다 바탕화면의 test.py 파이썬 파일 실행
