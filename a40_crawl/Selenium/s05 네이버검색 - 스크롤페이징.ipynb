#### 스크롤 끝까지 해보기

```python
query = '파이썬'
url = f'https://search.naver.com/search.naver?query={query}&nso=&where=blog&sm=tab_viw.all'

driver.get(url)

prev_height = driver.execute_script('return document.body.scrollHeight')

while True:
    # 끝까지 스크롤 다운
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    
    new_height = driver.execute_script('return document.body.scrollHeight')
    
    if new_height == prev_height: break  # 더 이상 스크롤 안되면 종료
        
    prev_height = new_height
```

```python
query = '파이썬'
url = f'https://search.naver.com/search.naver?query={query}&nso=&where=blog&sm=tab_viw.all'

driver.get(url)

prev_height = driver.execute_script('return document.body.scrollHeight')

while True:
    # 끝까지 스크롤 다운
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    
    # 2초 대기 : 페이징 로딩 되는 시간
    time.sleep(2)
    
    new_height = driver.execute_script('return document.body.scrollHeight')
    
    print(prev_height, new_height)
    
    if new_height == prev_height: break  # 더 이상 스크롤 안되면 종료
        
    prev_height = new_height
```

# Waits
https://www.selenium.dev/documentation/en/webdriver/waits/

- selenium 은 브라우저와 비동기로 작업하기 때문에 브라우저의 element 로딩될때(혹은 특정 조건이 될때)까지 기다린 다음에 수행해야 하는 코드가 있다면 Waits 객체를 활용해야 한다

#### 스크롤 n번만 하기

```python
driver.get('https://www.naver.com')
time.sleep(2)

query = '파이썬'
url = f'https://search.naver.com/search.naver?query={query}&nso=&where=blog&sm=tab_viw.all'

driver.get(url)
time.sleep(2)

prev_height = driver.execute_script('return document.body.scrollHeight')

for i in range(10):   
    # 끝까지 스크롤 다운
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    
    # 2초 대기 : 페이징 로딩 되는 시간
    time.sleep(2)
    
    new_height = driver.execute_script('return document.body.scrollHeight')
    
    print(prev_height, new_height)
    
    prev_height = new_height
```
