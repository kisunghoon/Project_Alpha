```python
import requests
from bs4 import BeautifulSoup
```

```python
def crawl_daum_blog(search, page = 1):  
    url = 'https://search.daum.net/search?nil_suggest=btn&w=blog&DA=PGD&q={search}&page={page}'.format(search = search, page = page)
    response = requests.get(url)
    dom = BeautifulSoup(response.text, "html.parser")
    
    post_elements = dom.select("#blogColl .list_info li .wrap_cont")


    result = [
        {
            'post_title' : post_element.select_one('.wrap_tit').text.strip(),
            'blog_title' : post_element.select_one('.info .f_nb').text.strip(),
            'url' : post_element.select_one('a').attrs.get('href').strip()
        }

        for post_element
        in post_elements
    ]

    return result
    
    
crawl_daum_blog('자바', 3)  # 동작 확인
```

```python
# 10페이지만 긁어오기

result = []
for page in range(1, 11):
    result += crawl_daum_blog('파이썬', page)
    
result
```
